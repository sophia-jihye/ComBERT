```python
import pandas as pd
import os, glob
```

```python
filepath = os.path.join('C:\DATA\ComBERT\data', 'Item1s_2016_by_sec_api_ExtractorApi_with_cik_of_sec_cik_mapper_3990.csv')
save_dir = './SEC-API.io'
if not os.path.exists(save_dir): os.makedirs(save_dir)
```

```python
df = pd.read_csv(filepath)
df
```

```python
for _, row in df.iterrows():
    company_name = 'Ticker={}_CIK={}_FiledAt={}'.format(row['ticker'], row['cik'], row['filedAt'])
    with open(os.path.join(save_dir, '{}.txt'.format(company_name)), 'w', encoding='utf8') as f:
        f.write(row['item_1'])

print('Saved {} files under {}'.format(len(glob.glob(os.path.join(save_dir, '*.txt'))), save_dir))
```
