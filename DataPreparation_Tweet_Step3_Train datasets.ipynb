```python
import pandas as pd 
import os
from tqdm import tqdm
tqdm.pandas()
from datetime import datetime
import re, html, copy
```

```python
train_full_filepath = os.path.join('C:\DATA\ComBERT\data_finetuning', 'train_261300.csv')

save_format = os.path.join('C:\DATA\ComBERT\data_finetuning', 'train_seed_{}', 'train_{}.csv')
```

## Data load

### Train dataset
* Produced in May, June, July, and August 2017

```python
train_full_df = pd.read_csv(train_full_filepath)
train_full_df
```

```python
for i in range(1, 2+1):
    for num_for_one_class in [100,150,200,250,300,350,400,450,500,1000,5000,10000,20000]:
        train_df = pd.concat([train_full_df[train_full_df['label']==label].sample(num_for_one_class, random_state=i) \
                              for label in train_full_df['label'].unique()])
        train_df = train_df.sample(frac=1).reset_index(drop=True) # shuffle
        
        filepath = save_format.format(i, len(train_df))
        if not os.path.exists(os.path.dirname(filepath)): os.makedirs(os.path.dirname(filepath))
        train_df.to_csv(filepath, index=False)
        print('Created {}'.format(filepath))
```
